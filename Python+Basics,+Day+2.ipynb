# Motion in a circle

```python
from math import pi, cos ,sin
from numpy import arange
```

```python

t = range(1,5)
for n in t :
    omega = 2 * pi/n
    #print(2 * pi/n)
    print(omega)
    #print(" ", omega)
    
```

```python
omega = 2.0
r = 1.0
t = arange(1,10,0.5)
for n in t:
    x = r*cos(omega*n)
    y = r*sin(omega*n)
    vx = -r*omega * sin(omega*n)
    vy = r*cos(omega*n)
    print(x,y,vx,vy)
    
    
```
